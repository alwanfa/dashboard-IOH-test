```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# data = pd.read_excel(
#     io="data/test_case.xlsx",
#     engine="openpyxl",
#     sheet_name='Data',
#     skiprows=1,
#     usecols = 'B:O')
data = pd.read_csv("data/test_case.csv")
```

```python
ro_code = ['BNRO', 'BRO', 'CJRO', 'CSRO', 'EJRO', 'JRO', 'KRO', 'MPRO','NSRO', 'SPRO', 'SSRO', 'WJRO']
dissolve_id = [ 6,  8,  1,  4,  2,  9,  3,  0,  7, 10,  5, 11]
map_ro_id = {}

for i, j in zip(ro_code, dissolve_id):
    map_ro_id[i] = j
```

```python
map_ro_id
```

```python
data['region_id'] = data['REGION'].apply(lambda x : map_ro_id[x])
```

```python
data.to_csv("data/test_case.csv")
```

```python
region_data = data.groupby(['REGION',"region_id"]).count().reset_index()
```

```python
data.drop('Unnamed: 0', axis = 1,inplace=True)
```

```python
data["Year"]= data["Year"].apply(lambda x : str(x))
```

```python
provider_data = data.groupby('Vendor').count().reset_index()
provider_data['Scope'] = provider_data['NE']
provider_data['OA %'] = provider_data['Actual']/provider_data['Scope']
provider_data['OA %'] = provider_data['OA %'].apply(lambda x: f"{(int(x*100))}%")
provider_data['Unsigned'] = provider_data['Scope']-provider_data["Plan"]
provider_data=provider_data[['Vendor','Scope','Plan','Actual', 'Unsigned', 'OA %']]
all = pd.DataFrame(provider_data.sum()).T[['Vendor','Scope','Plan','Actual', 'Unsigned', 'OA %']]
all['Vendor'] = "GRAND TOTAL"
# all.drop('Vedor', axis = 1, inplace = True)
all['OA %'] = f"{int((all.loc[0,'Actual']/all.loc[0,'Scope'])*100)}%"
final_df = pd.concat([provider_data, all], ignore_index=True).set_index('Vendor')
final_df
```

```python
years_data = data.groupby('Year').count().reset_index()
```

```python
# !pip install nbformat
fig = go.Figure(
    data =[
        go.Bar(name = "Actual", x = years_data["Year"].unique(), y =  years_data["Plan"]),
        go.Bar(name = "Actual", x = years_data["Year"].unique(), y =  years_data["Actual"])
    ], title = "Implemented Project"
)
fig.set_title()
fig.show()
```
